```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
import time
```

- Mengimpor BeautifulSoup soup dari modul Bs4 ke dalam kode pyton, BeautifulSoup merupakan pustaka python yang digunakan untuk mengurai (parsing) dokumen HTML atau XML dan mengestrak data, yang sangat berguna untuk web scraping.

- < import pandas as pd > :
- Perintah python yang digunakan untuk mengimpor pandas untuk menganalisis dan memanipulasi data. Pandas sendiri
adalah salah satu pustakas (library) eksternal yang bekerja dengan  data tabel seperti csv, excel, etc.

```python
nama_brand = []
harga_hp = []
total_terjual = []
toko_penjual = []
lokasi_penjual = []
rating_brand = []
```

- Pembuatan list [] kosong sementara bertujuan untuk menampung data-data hasil dari iterasi variable countainer_produk yang kemudian digunakan untuk mengisi data tiap kolom dan baris dari data frame. 
- Dalam python sendiri list merupakan array, yaitu struktur data yang tersusun terurut dari elemen yang didefinisikan dalam kurung siku. Dalam python list kosong didefinisikan sebagai list tanpa elemen atau item di dalam list.

```python
driver = webdriver.Safari()

for i in range (1,11):

    driver.get(f'https://www.bukalapak.com/c/handphone/hp-smartphone/hp-android?page={i}')
    inspect = driver.page_source
    time.sleep(1)
    html = driver.page_source

    soup = BeautifulSoup(html,'html.parser')

    container_produk = soup.find_all('div',{"class":"bl-flex-item mb-8"})

    for container in container_produk : 
        nama = container.find("p",{"class":"bl-text bl-text--body-14 bl-text--secondary bl-text--ellipsis__2"})
        if nama != None : 
            nama_brand.append(nama.text) 
        else:
            nama_brand.append(nama) 


        harga = container.find("p",{"class":"bl-text bl-text--semi-bold bl-text--ellipsis__1 bl-product-card-new__price"})
        if harga != None :
            harga_hp.append(harga.text)
        else:
            harga_hp.append(harga)


        lokasi = container.find("p",{"class":"bl-text bl-text--caption-12 bl-text--secondary bl-text--ellipsis__1 bl-product-card-new__store-location"})
        if lokasi != None :
            lokasi_penjual.append(lokasi.text)
        else:
            lokasi_penjual.append(lokasi)


        terjual = container.find("p",{"class":"bl-text bl-text--caption-12 bl-text--secondary bl-product-card-new__sold-count"})
        if   terjual != None :
            total_terjual.append(terjual.text)
        else:
            total_terjual.append(terjual)

        
        toko = container.find("p",{"class":"bl-text bl-text--caption-12 bl-text--secondary bl-text--ellipsis__1 bl-product-card-new__store-name"})
        if   toko != None :
            toko_penjual.append(toko.text)
        else:
            toko_penjual.append(toko)


        rating = container.find("p",{"class":"bl-text bl-text--caption-12 bl-text--bold"})
        if rating != None :
            rating_brand.append(rating.text)
        else:
            rating_brand.append(rating)


```

Dengan membuat urutan angka 1 - 11, dalam kasus ini yang dimaksud adalah melalui 1 - 10 halaman web, karena dalam python angka 11 akan menjadi tembok atau pembatas.

- < container_produk = soup.find_all('div',{"class":"bl-flex-item mb-8"}) > :

- container_produk adalah variabel berbentuk list untuk mewadahi semua bentuk elemen "produk" contoh: nama produk, harga, etc. yang  pada halaman web masih berbentuk kontainer penuh berisi data produk, berbentuk division < div > memiliki kelas "bl-flex-item mb-8", pada satu halaman web Bukalapak.com."


- < for container in container_produk > :

- Pengolahan variable container untuk meng-iterasi setiap elemen yang terdapat dalam container_produk, dimana berisikan daftar elemen yang didapat dari halaman web yang berisi informasi produk dalam container ('div',{"class":"bl-flex-item mb-8")."


 - < nama = container.find("p",{"class":"bl-text bl-text--body-14 bl-text--secondary bl-text--ellipsis__2"}) > :

- Kemudian menggunakan variable container dalam menjalankan pencarian single elemen di dalam {'class':"bl-text bl-text--body-14 bl-text--secondary bl-text--ellipsis__2"}, yang merupakan salah satu elemen dari container_produk. Dengan tujuan untuk menarik "nama_brand", yang kemudian dimasukan ke dalam variabel baru yaitu nama.


- < if nama != None > : 

- Sebuah kondisi yang dibuat sebagai filter, yaitu jika elemen dalam variabel nama ditemukan dan memiliki value, maka elemen tersebut tidak akan dianggap sebagai None.


- < nama_brand.append(nama.text) > :

- kondisi apabila elemen ditemukan dan memiliki value maka secara otomatis akan diekstrak kedalam bentuk text serta menghilangkan tag html nya. Lalu ditambahkan ke dalam list nama_brand.


- < else: nama_brand.append(nama) > :

- kondisi lainnya apabila tidak memiliki value akan di tambahkan ke dalam list nama_bran dengan bentuk None."

- Dilanjutkan dengan metode looping yang sama untuk list harga_hp hingga rating_brand

```python
data = pd.DataFrame({"Nama_Produk":nama_brand,
                    "Harga":harga_hp,
                    "Total_Terjual":total_terjual,
                     "Rating":rating_brand,
                     "Nama_Toko":toko_penjual,
                     "Kota_Toko":lokasi_penjual})
data
```

- < data = pd.DataFrame > :

- Merupakan tahap mengubah objek dari hasil web scraping website Bukalapak.com kedalam bentuk table lengkap yang terdiri dari index, baris dan kolom, atau yang disebut dengan Data Frame.

- Data Frame diatas memiliki data  index 0 - 99,  100 baris, dan 6 kolom 

```python
replacement = [(".",""),("\n",""),(" ","")]
for old, new in replacement:
    data['Harga'] = data['Harga'].str.replace(old, new)

replacement = [("\n","")]
for old, new in replacement:
    data['Nama_Produk'] = data['Nama_Produk'].str.replace(old, new)

    replacement = [("\n",""),("_"," ")]
for old, new in replacement:
    data['Nama_Toko'] = data['Nama_Toko'].str.replace(old, new)

    replacement = [("\n","")]
for old, new in replacement:
    data['Kota_Toko'] = data['Kota_Toko'].str.replace(old, new)

    replacement = [("\n",""),("Terjual","")]
for old, new in replacement:
    data['Total_Terjual'] = data['Total_Terjual'].str.replace(old, new)

    replacement = [("\n","")]
for old, new in replacement:
    data['Rating'] = data['Rating'].str.replace(old, new)


   
data
```

- < .str.replace() > :

- Definisi : Adalah metode dalam pandas untuk mengganti substring tertentu dalam nilai dari suatu kolom yang berisi data bertipe string yang berguna untuk memodifikasi data string dengan mengganti sebagian atau seluruh text yang terdapat pada setiap elemen.


- Berikut merupakan metode untuk  menghapus,mengganti symbol, koma, maupun menambahkan elemen untuk tipe data yang berbentuk string, dengan menggunakan metode .str.replace()

- < replacement = [(".",""),("\n",""),(" ","")] > :
- Dimulai dengan membuat variable replacement yang berisi elemen untuk menggantikan isi dari baris suatu kolom. Dalam kasus ini untuk menghilangkan simbol ".", "\n" dan spasi " ".


- < for old, new in replacement > :
- Sebuah perintah iterasi untuk data lama "old" dan data baru "new" dalam variable replacement.

- < data['Harga'] = data['Harga'].str.replace(old, new) > :
- Untuk kemudian mengubah isi data lama dalam kolom 'Harga' dengan data baru.
